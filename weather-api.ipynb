```python
import os
from dotenv import load_dotenv
load_dotenv()

API_KEY = os.getenv('PROJECT_API_KEY')
```

**Make a request for the current weather where you are born, or somewhere you've lived.**

```python
import requests
response = requests.get(f'https://api.weatherapi.com/v1/current.json?key={API_KEY}&q=manila')
data = response.json()
data
```

```python
data.keys()
```

**Print out the country this location is in.**

```python
data['location']['country']
```

**Print out the difference between the current temperature and how warm it feels. Use "It feels ___ degrees colder" or "It feels ___ degrees warmer," not negative numbers.**

```python
#Philippines temperature
manila_temp = data['current']['temp_f']
f'Temperature in Manila is {manila_temp} F'
```

```python
#Manila temperature real feel

manila_feels = data['current']['feelslike_f']
f'Temperature in Manila feels like {manila_feels} F'
```

```python
#Comparison

f'It feels {manila_feels - manila_temp} degrees warmer in Manila' 
```

**What's the current temperature at Heathrow International Airport? Use the airport's IATA code to search**

```python
#Heathrow Int'l Airport temperature IATA code is LHA

response = requests.get(f'http://api.weatherapi.com/v1/current.json?key={API_KEY}&q=iata:LHR')
data = response.json()
data
```

```python
lha_temp = data['current']['temp_f']

f'It is currently {lha_temp} F at Heathrow International Airport'
```

**What URL would I use to request a 3-day forecast at Heathrow?**


```python
#3 day forecast API request using location, days 
response = requests.get(f'http://api.weatherapi.com/v1/forecast.json?key={API_KEY}&q=iata:LHR&days=5')
data = response.json()

data
```

**Print the date of each of the 3 days you're getting a forecast for.**


```python
#3-day forecast
response = requests.get(f'http://api.weatherapi.com/v1/forecast.json?key={API_KEY}&q=iata:LHR&days=5')
data = response.json()

day_1 = data['forecast']['forecastday'][0]['date']
day_2 = data['forecast']['forecastday'][1]['date']
day_3 = data['forecast']['forecastday'][2]['date']
day_4 = data['forecast']['forecastday'][3]['date']

f'The days I am getting a forecast for are {day_2}, {day_3} and {day_4}'
```

**Print the maximum temperature of each of the days**

```python
#Maximum temperature of each days

response = requests.get(f'http://api.weatherapi.com/v1/forecast.json?key={API_KEY}&q=iata:LHR&days=5')
data = response.json()

max_day_1 = data['forecast']['forecastday'][1]['day']['maxtemp_f']
max_day_2 = data['forecast']['forecastday'][2]['day']['maxtemp_f']
max_day_3 = data['forecast']['forecastday'][3]['day']['maxtemp_f']

f'The maximum temperature for day 1 is {max_day_1} F, day 2 is {max_day_2} F, day 3 is {max_day_3} F,'

```
